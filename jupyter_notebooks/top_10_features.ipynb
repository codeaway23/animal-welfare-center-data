```python
# importing relevant packages. 

import pickle
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from xgboost.sklearn import XGBClassifier
```

```python
# preprating the data for training. 

train_data = pd.read_csv('../processed/train_data.csv')
train_ids = train_data['animal_id_outcome']
y = train_data['Label']
X = train_data.drop(['Label','animal_id_outcome'], axis=1)

test_data = pd.read_csv('../processed/test_data.csv')
test_ids = test_data['animal_id_outcome']

X_test = test_data.drop('animal_id_outcome', axis=1)

scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
scaled_X_test = scaler.transform(X_test)
```

```python
# plotting the correlation matrix heatmap for our training data

corrmat = train_data.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
g=sns.heatmap(train_data[top_corr_features].corr(),annot=False,cmap="RdYlGn")
g.get_figure().savefig('plots/corr_matrix_heatmap.png')
```

```python
## training the best classifier found from train_default.py + grid_search.py

model = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=3)
model.fit(scaled_X, y)
```

```python
# plotting the 10 most important features according to our trained classifier

fig = plt.figure()
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
fig.savefig('plots/xgboost_top_10_features.png')
```
